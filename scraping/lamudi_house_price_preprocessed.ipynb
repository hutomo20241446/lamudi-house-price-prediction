```python
import pandas as pd
data = pd.read_csv('lamudi_house_price_scraping.csv')

# Konversi kolom 'Price'  ke float
data['Price'] = data['Price'].str.replace('Rp', '')
data['Price'] = data['Price'].str.replace('.', '')
data['Price'] = data['Price'].astype(float)
data['Price'].info()
```

Fitur Price telah diubah menjadi numerikal (float64)
data

```python
import re
# Transformasi kolom 'Location'
def transform_location(location):
    # Ambil kata setelah koma terakhir
    last_part = location.split(',')[-1].strip()

    # Split kata-kata dalam bagian terakhir
    words = last_part.split()

    # Jika kata terakhir adalah satu atau dua kata tanpa angka, return seperti apa adanya
    if len(words) <= 2 and not any(re.search(r'\d', word) for word in words):
        return last_part
    else:
        # Sisakan hanya kata terakhir
        return words[-1]
# Transformasi pada kolom 'Location' ke nama kotanya
data.loc[:, 'Location'] = data['Location'].apply(transform_location)
data.groupby('Location').size().reset_index(name='Total_Rows')
```
